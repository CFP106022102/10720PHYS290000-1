```python
import numpy as np
import matplotlib.pyplot as plt 
import math

#SI unit
h = 6.62*10**-34  #普朗克常數
c=3*10**8         #光速
k=1.38*10**-23    #波茲曼常數
T=input("input temperature")
v = np.linspace(10**0, 10**15.5, 100, endpoint=True)
B=2*h*v**3/c**2/(math.e**(h*v/k/float(T))-1)

plt.plot(v,B,"g--")
```

```python
plt.loglog(v,B,"b")
```
