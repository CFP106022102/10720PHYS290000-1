```python
fp = open('Alice.txt','r',encoding='UTF-8')
import matplotlib.pyplot as plt
import numpy as np

line = fp.readline()
my_dict = {}

while line:
    s = line.split() 
    for x in s:
        if x not in my_dict:
            my_dict[x] = 1
        else:
            my_dict[x] += 1
    line = fp.readline()


fp.close()

num=[]

for key in my_dict:
    num.append(my_dict[key])
    
print(len(num))
num.sort()
num.reverse()

#線
plt.loglog(range(len(num)),num)
plt.show()

#柱狀
plt.hist(num,log=True,bins=1000)
plt.xscale('log')
plt.show()
```

```python
from scipy import optimize

def test_func(x, amp, alpha):
    return amp * x**alpha

params, params_covariance = optimize.curve_fit(test_func, range(1,len(num)+1), num)
print(params, params_covariance)

plt.loglog(range(1,len(num)+1), test_func(range(1,len(num)+1), params[0], params[1]), 'r*', label = 'Fitted function')



plt.show()
```
